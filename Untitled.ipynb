```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from time import sleep
import pickle
import sqlite3

def check_db_exists(con):
    cursor = con.cursor()
    query = """SELECT name FROM sqlite_master WHERE type='table' AND name='cars';"""
    with con:
        cursor.execute(query)
        if cursor.fetchone():
            return
    query = """
            CREATE TABLE IF NOT EXISTS cars
            (
                id          INTEGER NOT NULL PRIMARY KEY,
                name_reg         TEXT ,
                name_mark       TEXT,
                name_model     TEXT,
                pac TEXT,
                num    TEXT            
            );
            """
    with con:
        con.executescript(query)

def insert_flats_data_db(flats_data):
    con = sqlite3.connect('db.db')
    insert_flats_query = """
        INSERT OR IGNORE
        INTO cars(name_reg, name_mark, name_model, pac, num)
        VALUES (?, ?, ?, ?, ?);
        """
    flats_value = flats_data
    with con:
        con.execute(insert_flats_query, flats_value)

with open('file', 'rb') as f:
    a, b, c = pickle.load(f)

for delay in (10, 15, 700, 1200):
    try:
        driver = webdriver.Chrome()
        driver.maximize_window()
        driver.get("https://kolesa.kz/")
        
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "(//span[@class='action-link'])[1]"))).click()
        
        # Расширенный поиск
        sleep(2)
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//span[contains(text(), 'Расширенный поиск')]")))
        driver.find_element(By.XPATH, "//button[@class='btn set-optional']").click()
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//span[contains(text(), 'В наличии')]")))
        sleep(2)
        label = driver.find_element(By.CSS_SELECTOR, 'label[for="auto-car-order-1"]')
        driver.execute_script("arguments[0].click();", label)
        sleep(1)
        driver.find_element(By.XPATH, "//button[@class='btn set-optional']").click()
        sleep(1)
        # еще
        WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'button.FilterItem.FilterGroup__toggle.isHighlighted'))).click()
        
        name = "div[class='filter-region__body'] button"
        citis = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, name)))
        
        con = sqlite3.connect('db.db')
        check_db_exists(con)
        for ii in range(a, len(citis)):
            city = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, name)))[ii]
            name_reg = city.text
            city.click()
            sleep(1)
            # еще
            WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "(//span[contains(text(), 'ещё')])[2]"))).click()
            sleep(0.5)
            name1 = "div[class='grouped-list filter-car__extended-group'] button"
            marks = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, name1)))
            sleep(0.5)
            WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "(//span[contains(text(), 'ещё')])[2]"))).click()
            sleep(0.5)
            for ma in range(b, len(marks)):
                WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "(//span[contains(text(), 'ещё')])[2]"))).click()
                sleep(0.5)
                mark = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, name1)))
                name_mark = mark[ma].text
                mark[ma].click()
                sleep(0.5)
                # еще
                WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "(//span[contains(text(), 'ещё')])[3]"))).click()
                sleep(0.5)
                models = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, name1)))
                WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "(//span[contains(text(), 'ещё')])[3]"))).click()
                sleep(0.5)
                for m in range(c, len(models)):
                    WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "(//span[contains(text(), 'ещё')])[3]"))).click()
                    sleep(0.3)
                    model = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, name1)))
                    name_model = model[m].text
                    model[m].click()
                    sleep(0.3)
                    num_temp = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//span[@class='js__search-form-submit-value']"))).text.strip()
                    if num_temp == 'Ничего не найдено':
                        continue
                    # все
                    WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//span[contains(text(), 'все')]"))).click()
                    sleep(0.3)
                    checkboxes = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.XPATH, "//div[@class='generation']")))
                    for cb in checkboxes:
                        driver.execute_script("arguments[0].scrollIntoView({block: 'center'});", cb)
                        sleep(1)
                        if not cb.is_selected():
                            pac = cb.text
                            cb.click()
                            sleep(0.5)
                            num = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//span[@class='js__search-form-submit-value']"))).text
                            cb.click()
                            sleep(0.3)
                            insert_flats_data_db([name_reg, name_mark, name_model, pac, num])
            WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//span[contains(text(), 'Все регионы')]"))).click()
    except Exception as error:
        print(error)
        a,b,c = ii,ma,m 

    finally: 
        driver.close()
        with open('file', 'wb') as f:
            pickle.dump((ii,ma,m), f)
    sleep(delay)
```
